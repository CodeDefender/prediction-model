```
import numpy as np
from sklearn.neighbors.kde import KernelDensity
```

```
Arr = np.array([[1,2,4],[1,3,6],[4,5,6],[45,50,55000],[10,10,10]])
```

```
kde = KernelDensity(kernel='gaussian',bandwidth=0.5).fit(Arr)
y = kde.score_samples([[1,2,5]])
np.exp(y)
```

```
y = kde.score_samples(Arr)
np.exp(y)
```
