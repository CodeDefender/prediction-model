```
from google.colab import drive
drive.mount('/content/gdrive')
from sklearn import datasets 
from sklearn.metrics import confusion_matrix 
from sklearn.model_selection import train_test_split 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
X_11 = pd.read_csv("/content/gdrive/My Drive/cloud_data/data/d11.csv")
y_9 = pd.read_csv("/content/gdrive/My Drive/cloud_data/data/d9.csv")
X_10 = pd.read_csv("/content/gdrive/My Drive/cloud_data/data/d10.csv")
y_8 = pd.read_csv("/content/gdrive/My Drive/cloud_data/data/d8.csv")
```

```
cleanup_nums = {"Class Name":{"'High'": 4, "'Medium'": 3,"'Low'": 2,"'Very Low'": 1}}
X_10.replace(cleanup_nums,inplace=True)
y_8.replace(cleanup_nums,inplace=True)

cleanup_nums = {"Class Name":{"'High'": 4, "'Very High'": 3,"'Low'": 2,"'Very Low'": 1}}
X_11.replace(cleanup_nums,inplace=True)
y_9.replace(cleanup_nums,inplace=True)
```

```
used_features =[
    
   "CPU Utilization",
    "Response Time"
]
```

```
from sklearn.tree import DecisionTreeClassifier 
import numpy as np

knn_cpu = list()
knn_response = list()
for i in range(2,26):
   
    knn_cpu.append(DecisionTreeClassifier(max_depth = i).fit( X_10[used_features], X_10["Class Name"])) 
    knn_response.append(DecisionTreeClassifier(max_depth = i).fit( X_11[used_features], X_11["Class Name"])) 

# accuracy on X_test 

accuracy_cpu=list()
accuracy_response=list()
for i in range(24):
  
  accuracy_cpu.append(knn_cpu[i].score(y_8[used_features], y_8["Class Name"]))
  accuracy_response.append(knn_response[i].score(y_9[used_features], y_9["Class Name"]))
#print(accuracy_memory)
#print(accuracy_cpu)

print(max(accuracy_cpu))
print(max(accuracy_response))

print((accuracy_cpu[5]))
print((accuracy_response[12]))


accuracy_cpu = np.array(accuracy_cpu)
accuracy_cpu = accuracy_cpu*100
accuracy_response = np.array(accuracy_response)
accuracy_response = accuracy_response*100

#plt.figure().suptitle('KNN Model-2', fontsize=12, fontweight='bold')

plt.plot(np.arange(2,26),accuracy_cpu,'b-')
plt.plot(np.arange(2,26),accuracy_response,'r-')
plt.legend(["CPU Utilization","Response Time"])
#plt.plot(7,accuracy_cpu[5],'bo',14,accuracy_response[12],'ro')
plt.ylabel("Accuracy of Model")
plt.xlabel("Depth size")
plt.title("Decision Tree Model-2",fontsize=12, fontweight='bold')
plt.show()
```
